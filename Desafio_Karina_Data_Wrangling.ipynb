```python
import pandas as pd

# Load the Excel files
file_aula0 = 'C:/Users/karin/OneDrive/Área de Trabalho/data science/aula0.xlsx'
file_aula1 = 'C:/Users/karin/OneDrive/Área de Trabalho/data science/aula1.xlsx'

# Reading all sheets from both Excel files
data_aula0 = pd.read_excel(file_aula0, sheet_name=None)
data_aula1 = pd.read_excel(file_aula1, sheet_name=None)

# Display the sheet names and the first few rows of each sheet to understand the structure
list_aula0 = {sheet: data_aula0[sheet].head() for sheet in data_aula0}
list_aula1 = {sheet: data_aula1[sheet].head() for sheet in data_aula1}

list_aula0, list_aula1
```

## Merge

```python
# Let's perform the merge operations (inner, outer, left, right) on the two dataframes.

df_aula0 = data_aula0['Sheet1']
df_aula1 = data_aula1['Sheet1']

# Performing merge operations on the columns key1 and key2
merge_inner = pd.merge(df_aula0, df_aula1, on=['key1', 'key2'], how='inner')
merge_outer = pd.merge(df_aula0, df_aula1, on=['key1', 'key2'], how='outer')
merge_left = pd.merge(df_aula0, df_aula1, on=['key1', 'key2'], how='left')
merge_right = pd.merge(df_aula0, df_aula1, on=['key1', 'key2'], how='right')
```

```python
# Now let's also concatenate the two dataframes along rows and columns
concat_axis0 = pd.concat([df_aula0, df_aula1], axis=0, ignore_index=True)
concat_axis1 = pd.concat([df_aula0, df_aula1], axis=1)
```
